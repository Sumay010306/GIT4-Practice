```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import os
folder_name = r"C:/Sumay"
name = "500hits.csv"
print(os.path.join(folder_name,name))
```

```python
data = pd.read_csv('C:/Sumay/500hits.csv', encoding='latin-1')
data.head()
```

```python
data
```

```python
X= data.drop(columns=['PLAYER', 'HOF'])
```

```python
X=(X-X.mean())/X.std()
```

```python
y= data['HOF']
```

```python
y=(y-y.mean())/y.std()
```
